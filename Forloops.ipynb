#loops in python

```python
i = 1          # initializing
while i<=5:    # condition
    print('data science')
    i = i + 1  # increment
```

```python
i = 5          # initializing

while i>=1:    # condition
    print('data science')
    i = i - 1  # decrement
```

```python
i = 1

while i<=5:
    print('data science') # when we mention end then new line will not create
    j = 1
    while j<=4:
        print('technology')
        j = j + 1
        
    i = i + 1
    print()
    
    # the output which we got is very lengty but how to make them one line lets refer to below code
```

```python
 i = 1
while i<=5:
    print(' datascience', end = "") # when we mention end then new line will not create
    j = 1
    while j<=4:
        print(' technology', end="")
        j = j + 1
        
    i = i + 1
    print()
```
